### In this problem, you'll write a function called evenTuples!

So if test is the tuple ('I', 'am', 'a', 'test', 'tuple'), then evaluating evenTuples on this input would return the tuple ('am', 'test').

a new tuple with even elements<br>

```python
# Problem 1
# Name: Your name
# Time: 3:30 
# ... your code goes here ... 

def evenTuples(aTup):
    '''
    aTup: a tuple
    
    returns: tuple, every even other element of aTup. 
    '''
    # Your Code Here
    
    if type(aTup)==tuple:
        return (aTup[1::2])
 
aTup = ('I', 'am', 'a', 'test', 'tuple')
print(evenTuples(aTup))
```

# Problem 2. Calculate the mean and median (計算平均數與中位數)

The mean and median are commonly used values in statistics. Design a function, enter a sequence of numbers (list), and return its mean and median. The average is equal to the sum of all values divided by the number; the median refers to a group of the data that is sorted from small to large, and the middle number is selected. If there are even numbers of observations, the median is not unique. Usually, the average of the two middle values is taken as the median. The input and output instructions are as follows:<br>
Your code must be iterative - use of the sum(), median(), and mean() is not allowed.<br>

Input description:<br>
A sequence of numbers: \[10, 20, 30, 30, 30, 40\]<br>
Output description:<br>

Median: 30<br>

平均數與中位數是統計學中常用的數值。設計一個方程式，輸入一序列數字(list)，並回傳其平均數(mean)與中位數(median)，平均值等於所有值累加後除以其個數；中位數是指將一組資料從小到大排序後，取最中間的數，如果觀察值有偶數個，則中位數不唯一，通常取最中間的兩個數值的平均數作爲中位數。輸出入說明如下：<br>
你的程式碼必須是迭代的-不允許使用sum(), median(), mean()等函式。<br>

輸入說明：<br>
一序列數字：\[10, 20, 30, 30, 30, 40\]<br>
輸出說明：<br>
平均數: 26.66<br>
中位數: 30<br>

```python
# Problem 2
# Name: Your name
# Time: 3:30 
# ... your code goes here ... 

def describe(L):
    """
    """
    s=0
    L.sort()
    for k in L:
        s+=k
    s/=len(L)
    m=0
    f=len(L)
    if len(L)%2==0:
        m=(L[f//2]+L[f//2-1])/2
    else:
        m=L[(f-1)//2]
    return s,m
L=[1, 2, 3, 4, 5, 6]
(mea, med)= describe(L)
print("平均數: ", mea)
print("中位數: ", med)

```

# Problem 3：ISBN-10 Check (國際標準書號 ISBN-10檢查)
Design an function, enter an International Standard Book Number ISBN, and return whether the ISBN code is correct or not. For example, a complete ISBN-10 International Standard Book Number is ISBN 7-309-04547-6, the last digit is the check code, the numbers 0 to 9, or the letter X (representing 10), which is used to verify whether the ISBN code is correct or not. . <br>
<br>
The calculation method of the 10-digit check code is to multiply the first 9 digits by the weights of \[10, 9, 8, 7, 6, 5, 4, 3, 2] and add them, and divide this sum by 11 to find the remainder. It is the check code. <br>

For example: Assume that the first 9 digits of an international standard book number are: 7-309-04547. <br>
<br>
Calculate the weighted sum: 7\*10 + 3\*9 + 0\*8 + 9\*7 + 0\*6 + 4\*5 + 5\*4 + 4\*3 + 7\*2 = 226 <br>
The check code is the remainder of the weighted sum divided by 11 226=20\*11 ... 6<br>
If the check code is 10, it needs to be converted into a Roman numeral "X"<br>
<br>
Therefore, the check code of ISBN 7-309-04547-6 is 6, and the calculated check code is 6, and return True<br>
In addition, the check code of ISBN 9-620-01006-X is X, and the calculated check code is 1, return False<br>
設計一個方程式，輸入一國際標準書號 ISBN，並回傳該ISBN碼是否正確屬實。例如一完整的ISBN-10國際標準書號為ISBN 7-309-04547-6，最後一位為校驗碼，數字0到9，或字母X（代表10），作用為驗證該ISBN碼是否正確屬實。<br>
<br>
10 位校驗碼的計算方法為前 9 位分別乘上\[10, 9, 8, 7, 6, 5, 4, 3, 2]加權權數後相加，將此和除以 11 求餘數，就是校驗碼。<br>

例子：假設某國際標準書號的前 9 位是：7-309-04547。<br>
<br>
計算加權和：7\*10 + 3\*9 + 0\*8 + 9\*7 + 0\*6 + 4\*5 + 5\*4 + 4\*3 + 7\*2 = 226<br>
校驗碼即為加權和除以 11 的餘數 226=20\*11 ... 6<br>
如果校驗碼是 10，需轉成羅馬數字「X」<br>
<br>
所以，ISBN 7-309-04547-6 的校驗碼是 6，計算出之校驗碼是 6，回傳True<br>
另外，ISBN 9-620-01006-X 的校驗碼是 X，計算出之校驗碼是 1，回傳False<br>

<br>
Input<br>
<序列> <br>
Output<br>
<序列平均值> <br>
<br>
Sample Input<br>
ISBN 7-309-04547-6<br>
<br>
Sample Output<br>
True<br>

Sample Input<br>
ISBN 9-620-01006-X<br>

Sample Output<br>
False<br>

```python
# Problem 3
# Name: Your name
# Time: 3:30 
# ... your code goes here ... 

def isbn_check(isbn):
    """
    """
    isbn2=isbn[5:-1]
    isbn2=list(isbn2)
    for x in isbn2:
        if x=="-":
            isbn2.remove("-")
    isbn2=list(map(int,isbn2))
    check=[10, 9, 8, 7, 6, 5, 4, 3, 2]
    sum=0
    f=0
    for x in isbn2[:]:
        sum+=x*check[f]
        f+=1
    cnum=sum%11
    id=isbn[-1]
    if id=="X":
        id=10
    id=int(id)
    if cnum==id:
        return True
    else:
        return False

print(isbn_check("ISBN 7-309-04547-6"))
```
